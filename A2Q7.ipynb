## 1. Determine the value of the maximum time $b$

In a PERT distribution, the expectation is calculated using the formula:
$$E[X] = \frac{a + 4c + b}{6}$$

The problem states that the expectation $E[X]$ must match the most likely time $c$. Therefore, we set $E[X] = c$:
$$c = \frac{a + 4c + b}{6}$$

Now, solve for $b$:
1. Multiply both sides by 6:
   $$6c = a + 4c + b$$
2. Subtract $4c$ from both sides:
   $$2c = a + b$$
3. Isolate $b$:
   $$b = 2c - a$$

Substitute the given values $a = 5$ and $c = 11$:
$$b = 2(11) - 5$$
$$b = 22 - 5$$
$$\mathbf{b = 17 \text{ days}}$$

---

## 2. What happens to the variance?

When $E[X] = c$, the distribution becomes **perfectly symmetric**.

* **Symmetry:** Because $b - c = 17 - 11 = 6$ and $c - a = 11 - 5 = 6$, the mode is exactly in the center of the range.

* **Shape Parameters:** This symmetry implies that $\alpha = \beta$. Using the PERT formulas:
  $$\alpha = 1 + 4\left(\frac{11-5}{17-5}\right) = 1 + 4(0.5) = 3$$
  $$\beta = 1 + 4\left(\frac{17-11}{17-5}\right) = 1 + 4(0.5) = 3$$


* **Variance Impact:** The variance is minimized for a fixed range $[a, b]$ when the distribution is symmetric. Using the approximation provided in the previous question:
  $$Var(X) = \frac{(11-5)(17-11)}{7} = \frac{6 \cdot 6}{7} = \frac{36}{7} \approx 5.14$$


If the distribution were skewed (if $b$ were much larger), the variance would increase because the "tail" of the distribution would be longer, creating more uncertainty. In this specific case, matching the expectation to the mode indicates a **stable and balanced risk profile** where delays and early finishes are equally likely.
