## 10: Information Theory Calculations

We will now compute key information theory metrics to quantify the uncertainty regarding fraudulent transactions and how knowing the customer's risk class ($Y$) reduces this uncertainty.

### a) Compute Entropy $H(X)$ (Initial Uncertainty)

Given $P(X=1) = 0.30$ and $P(X=0) = 0.70$:

$$H(X) = -P(X=0) \log_2(P(X=0)) - P(X=1) \log_2(P(X=1))$$
$$H(X) = -0.70 \log_2(0.70) - 0.30 \log_2(0.30)$$
$$H(X) \approx 0.8813 \text{ bits}$$

### b) Compute Conditional Entropy $H(X \mid Y)$ (Remaining Uncertainty)

This represents the uncertainty about $X$ after knowing $Y$. It is calculated as a weighted average of conditional entropies:

$$H(X \mid Y) = P(Y=0)H(X \mid Y=0) + P(Y=1)H(X \mid Y=1)$$

#### For $Y=0$:

Using $p_{X|Y}(0 \mid 0) = 5/6$ and $p_{X|Y}(1 \mid 0) = 1/6$:

$$H(X \mid Y=0) = -\frac{5}{6}\log_2\left(\frac{5}{6}\right) - \frac{1}{6}\log_2\left(\frac{1}{6}\right) \approx 0.6500 \text{ bits}$$

#### For $Y=1$:

Using $p_{X|Y}(0 \mid 1) = 1/2$ and $p_{X|Y}(1 \mid 1) = 1/2$:

$$H(X \mid Y=1) = -\frac{1}{2}\log_2\left(\frac{1}{2}
ight) - \frac{1}{2}\log_2\left(\frac{1}{2}
ight) = 1.0000 \text{ bit}$$

#### Combining them:

$$H(X \mid Y) = 0.60(0.6500) + 0.40(1.0000) = 0.3900 + 0.4000 = 0.7900 \text{ bits}$$

### c) Calculate Mutual Information $I(X; Y)$

Mutual Information quantifies the amount of information obtained about one random variable by observing the other. It is the reduction in uncertainty of $X$ due to knowing $Y$:

$$I(X; Y) = H(X) - H(X \mid Y)$$
$$I(X; Y) = 0.8813 - 0.7900 = 0.0913 \text{ bits}$$

# Q2) Conditional Density of the Multivariate Gaussian

### 1. Theoretical Foundation & Block Matrix Identities

To derive the conditional density $f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the fundamental definition of conditional density:

$$f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

Since the joint distribution is Gaussian, the marginal distribution $\mathbf{Y}$ is also Gaussian: $\mathbf{Y} \sim \mathcal{N}(\boldsymbol{\mu}_Y, \boldsymbol{\Sigma}_{YY})$.

To compute this fraction efficiently, we must handle the matrix inverse $\boldsymbol{\Sigma}^{-1}$ and the determinant $|\boldsymbol{\Sigma}|$ using Schur Complements.

#### Block Matrix Factorization

The joint covariance matrix can be factored using the Schur complement of $\boldsymbol{\Sigma}_{YY}$, denoted as $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$:

$$\begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix} = \begin{bmatrix} \mathbf{I}_n & \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I}_m \end{bmatrix} \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY} \end{bmatrix} \begin{bmatrix} \mathbf{I}_n & \mathbf{0} \\ \boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} & \mathbf{I}_m \end{bmatrix}$$

#### Determinant Identity

Taking the determinant of both sides of the factorization yields:

$$|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{X\mid Y}| \cdot |\boldsymbol{\Sigma}_{YY}|$$

#### Inverse Identity (Block Matrix Inversion)

Inverting the block components yields the expression for $\boldsymbol{\Sigma}^{-1}$:

$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \mathbf{I}_n & \mathbf{0} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} & \mathbf{I}_m \end{bmatrix} \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{I}_n & -\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I}_m \end{bmatrix}$$

Performing the outer multiplications gives:

$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & -\boldsymbol{\Sigma}_{X\mid Y}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{X\mid Y}^{-1} & \boldsymbol{\Sigma}_{YY}^{-1} + \boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{X\mid Y}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix}$$

### 2. Algebraic Transformation of the Quadratic Form

Let us define the centered vectors:

$$\mathbf{z}_x = \mathbf{x} - \boldsymbol{\mu}_X, \qquad \mathbf{z}_y = \mathbf{y} - \boldsymbol{\mu}_Y$$

The quadratic form in the joint exponent can be rewritten using the factored structure of $\boldsymbol{\Sigma}^{-1}$:

$$(\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) = \begin{bmatrix} \mathbf{z}_x^T & \mathbf{z}_y^T \end{bmatrix} \begin{bmatrix} \mathbf{I}_n & \mathbf{0} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} & \mathbf{I}_m \end{bmatrix} \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{I}_n & -\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I}_m \end{bmatrix} \begin{bmatrix} \mathbf{z}_x \\ \mathbf{z}_y \end{bmatrix}$$

Let's compute the rightmost vector multiplication first:

$$\begin{bmatrix} \mathbf{I}_n & -\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I}_m \end{bmatrix} \begin{bmatrix} \mathbf{z}_x \\ \mathbf{z}_y \end{bmatrix} = \begin{bmatrix} \mathbf{z}_x - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{z}_y \\ \mathbf{z}_y \end{bmatrix}$$

By symmetry, the leftmost transformation yields:

$$\begin{bmatrix} \mathbf{z}_x^T & \mathbf{z}_y^T \end{bmatrix} \begin{bmatrix} \mathbf{I}_n & \mathbf{0} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} & \mathbf{I}_m \end{bmatrix} = \begin{bmatrix} \left(\mathbf{z}_x - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{z}_y\right)^T & \mathbf{z}_y^T \end{bmatrix}$$

Substituting these back into the core diagonal matrix equation simplifies the entire quadratic expression into two independent parts:

$$(\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) = \left(\mathbf{z}_x - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{z}_y\right)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} \left(\mathbf{z}_x - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{z}_y\right) + \mathbf{z}_y^T \boldsymbol{\Sigma}_{YY}^{-1} \mathbf{z}_y$$

Notice that the first term relies heavily on the conditional mean formula:

$$\mathbf{z}_x - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\mathbf{z}_y = (\mathbf{x} - \boldsymbol{\mu}_X) - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y) = \mathbf{x} - \boldsymbol{\mu}^*$$

Thus, the final joint quadratic form is:

$$(\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) = (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) + (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y)$$

### 3. Constructing the Conditional Density Ratio

Now we substitute our derived determinant and exponent identities directly back into the ratio definition:

$$f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}_{X\mid Y}| |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)}{\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)}$$

#### Term Cancellation

The marginal density terms involving $\mathbf{y}$ inside the exponential functions cancel out perfectly.
The determinant parameter $|\boldsymbol{\Sigma}_{YY}|^{1/2}$ cancels from the denominator.
The scalar base scaling terms reduce neatly: $\frac{(2\pi)^{m/2}}{(2\pi)^{(n+m)/2}} = \frac{1}{(2\pi)^{n/2}}$.

After final reductions, we obtain the clean target density expression:

$$f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X\mid Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

where

$\boldsymbol{\Sigma}_{X\mid Y} = \mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$

$\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$

This proves that the conditional distribution $\mathbf{X} \mid \mathbf{Y} = \mathbf{y}$ is a multivariate Gaussian distribution:

$$\mathbf{X} \mid \mathbf{Y} = \mathbf{y} \sim \mathcal{N}(\boldsymbol{\mu}^*, \boldsymbol{\Sigma}_{X\mid Y})$$

# Q3) Continuous Fault Monitoring of a Spindle Motor



## 1. Joint Distribution of $(X, Y)$

* **Distribution of $Y$:** We are given $Y = X + Q$, where $X \sim \mathscr{N}(0, 1)$ and $Q \sim \mathscr{N}(0, 4)$ are independent.
  $$\mu_Y = E[Y] = E[X + Q] = E[X] + E[Q] = 0 + 0 = 0$$
  $$\sigma_Y^2 = \text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$$
  Since $X$ and $Q$ are independent Gaussians, their sum is also Gaussian:
  $$Y \sim \mathscr{N}(0, 5)$$

* **Covariance $\text{Cov}(X, Y)$:**
  $$\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Cov}(X, X) + \text{Cov}(X, Q)$$
  Since $\text{Cov}(X, X) = \text{Var}(X) = 1$ and $\text{Cov}(X, Q) = 0$ (due to independence):
  $$\text{Cov}(X, Y) = 1 + 0 = 1$$

* **Joint Matrix Form:**
  The random vector $\mathbf{Z} = \begin{bmatrix} X \\ Y \end{bmatrix}$ follows a bivariate normal distribution:
  $$\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathscr{N}\left( \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix} \right)$$

---

## 2. Information Geometry & Measurability

$X$ is **not** $\sigma(Y)$-measurable because observing $Y$ does not uniquely determine $X$. For any observed value $Y = y$, there are infinite pairs of $(X, Q)$ such that $X + Q = y$.

In practical terms, if the power meter reads $Y = 5$:
* It could mean the motor is heavily overheating ($X = 5$) while running under zero cutting load ($Q = 0$).
* It could mean the motor is perfectly cool ($X = 0$) but cutting a highly resistant, hard material ($Q = 5$).
* It could be a blend of both ($X = 1, Q = 4$).

Because the single sensor reading combines two distinct physical phenomena, the engineer encounters an **identifiability problem**. The power meter alone cannot isolate the source of the thermal vs. mechanical stress.

---

## 3. Conditional Expectation (The Best Estimate)

Substituting our values ($\mu_X = 0$, $\mu_Y = 0$, $\text{Cov}(X,Y) = 1$, $\text{Var}(Y) = 5$) into the formula:
$$E[X \mid Y = y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$

Thus, our optimal estimator function is:
$$\widehat{X} = E[X \mid Y] = \frac{1}{5}Y$$

### **Signal-to-Total-Variance Ratio Interpretation**
We can rewrite this weight by separating the variance components:
$$\text{Weight} = \frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)} = \frac{1}{1 + 4} = \frac{1}{5}$$
This fraction acts as a **signal-to-total-variance filter** (analogous to a static Wiener filter coefficient). It measures how "trustworthy" the power meter is as a proxy for temperature. Because 80% of the variance comes from mechanical noise ($Q$), we only trust 20% ($\frac{1}{5}$) of the scale of the power reading when predicting temperature.

---

## 4. Verification of the Residual


* **Proof that $E[Z] = 0$:**
  $$E[Z] = E[X - \widehat{X}] = E\left[X - \frac{1}{5}Y\right] = E[X] - \frac{1}{5}E[Y] = 0 - \frac{1}{5}(0) = 0 \quad \checkmark$$

* **Verification of Orthogonality ($\text{Cov}(Z, Y) = 0$):**
  $$\text{Cov}(Z, Y) = \text{Cov}\left(X - \frac{1}{5}Y, Y\right) = \text{Cov}(X, Y) - \frac{1}{5}\text{Cov}(Y, Y)$$
  Since $\text{Cov}(X, Y) = 1$ and $\text{Cov}(Y, Y) = \text{Var}(Y) = 5$:
  $$\text{Cov}(Z, Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0 \quad \checkmark$$

Because the covariance is zero, the error residual $Z$ is uncorrelated with the sensor observations $Y$.

---

## 5. Information Theory: Differential Entropy


* **Baseline Differential Entropy $h(X)$:**
  Since $\sigma_X^2 = 1$:
  $$h(X) = \frac{1}{2} \log_2(2\pi e (1)) = \frac{1}{2} \log_2(2\pi e) \approx 2.0471 \text{ bits}$$

* **Conditional Differential Entropy $h(X \mid Y)$:**
  First, calculate the remaining conditional variance $\sigma_{X|Y}^2$:
  $$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{5} = 1 - 0.20 = 0.80$$
  Now evaluate the entropy equation with this remaining variance:
  $$h(X \mid Y) = \frac{1}{2} \log_2(2\pi e \sigma_{X|Y}^2) = \frac{1}{2} \log_2(2\pi e \cdot 0.80) \approx 1.8861 \text{ bits}$$

---

## 6. Mutual Information


* **Calculation:**
  $$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2} \log_2(2\pi e) - \frac{1}{2} \log_2(2\pi e \cdot 0.80)$$
  Using logarithmic division rules:
  $$I(X; Y) = \frac{1}{2} \log_2\left(\frac{2\pi e}{2\pi e \cdot 0.80}\right) = \frac{1}{2} \log_2\left(\frac{1}{0.80}\right) = \frac{1}{2} \log_2(1.25) \approx 0.1610 \text{ bits}$$

* **Uncertainty Reduction Percentage:**
  To evaluate how much uncertainty is removed, we divide the mutual information by the baseline entropy:
  $$\text{Percentage} = \frac{I(X; Y)}{h(X)} \times 100\% = \frac{0.1610}{2.0471} \times 100\% \approx 7.86\%$$
  Observing the power meter removes only about **7.86%** of the baseline uncertainty regarding the internal spindle temperature.

* **Scenario Change (High Noise: $\text{Var}(Q) = 100$):**
  If mechanical noise surges to $100$:
  $$\text{Var}(Y_{\text{new}}) = 1 + 100 = 101, \qquad \text{Cov}(X, Y_{\text{new}}) = 1$$
  The conditional variance becomes:
  $$\sigma_{X|Y_{\text{new}}}^2 = 1 - \frac{1}{101} = \frac{100}{101} \approx 0.9901$$
  Recalculating Mutual Information:
  $$I(X; Y_{\text{new}}) = \frac{1}{2} \log_2\left(\frac{1}{0.9901}\right) = \frac{1}{2} \log_2(1.01) \approx 0.0072 \text{ bits}$$
  $$\text{Percentage}_{\text{new}} = \frac{0.0072}{2.0471} \times 100\% \approx 0.35\%$$
  As mechanical noise spikes, the power sensor becomes overwhelmed, reducing the mutual information to near zero.

## 7. Diagnostic Interpretation & Shrinkage

### **Analysis of the Manager's Claim**
The manager suggests using $\widehat{X}_{\text{manager}} = Y$ as the estimate. Let's calculate the expected value of this estimator given a true temperature state $X$:
$$E[\widehat{X}_{\text{manager}} \mid X] = E[X + Q \mid X] = X + E[Q] = X$$
While this shows the estimator is *unbiased on average* over long periods, it is an unstable choice for single-instance diagnostics because it ignores the risk of noise.

If we look at it from the perspective of conditional tracking given an observation:
$$E[Y \mid Y=y] = y \neq \frac{1}{5}y$$

### **Why the Optimal Estimator "Shrinks" Toward Zero**
Our mathematically optimal estimator $\widehat{X} = \frac{1}{5}Y$ intentionally **shrinks** the raw measurement toward the global prior mean ($0$).

This shrinkage happens because the model balances sensor reading information with historical prior knowledge.
* When the power meter reads $Y = 5$, the math recognizes that most of this jump is likely caused by the high-variance mechanical noise variable ($Q \sim \mathscr{N}(0,4)$) rather than the lower-variance temperature signal ($X \sim \mathscr{N}(0,1)$).
* To minimize the total squared error (MSE), the estimator scales the reading down by a factor of 5.

Ignoring this calculation and assuming $X = Y$ will regularly cause the factory to overreact to normal cutting torque variations, mistaking regular mechanical workloads for catastrophic engine overheating.
