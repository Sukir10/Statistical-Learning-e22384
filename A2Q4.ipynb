## 1. Construct a Probability Space $(\Omega, \mathfrak{F}, P)$

* **Sample Space ($\Omega$):** Since $X$ represents a proportion, the set of all possible outcomes is the unit interval:
   
   $$\Omega = [0, 1]$$

* **$\sigma$-algebra ($\mathfrak{F}$):** We use the Borel $\sigma$-algebra on $[0, 1]$, denoted as $\mathcal{B}([0, 1])$, which contains all intervals and their countable unions/intersections.
* **Probability Measure ($P$):** For any event $A \in \mathfrak{F}$, the measure is defined by the integral of the density function:

    $$P(A) = \int_A f(x) \, dx$$

---

## 2. Define $X$ as a continuous random variable

$X$ is the identity mapping $X: \Omega \to [0, 1]$ such that $X(\omega) = \omega$. It is a continuous random variable because its distribution can be described by a density function $f(x)$ such that for any $a, b \in [0, 1]$:
$$P(a \le X \le b) = \int_a^b f(x) \, dx$$

---

## 3. Probability Density Function (PDF)

The PDF of $X \sim \text{Beta}(2, 18)$ is:
$$f(x) = \begin{cases} \frac{1}{B(2, 18)}x(1-x)^{17}, & 0 < x < 1 \\ 0, & \text{otherwise} \end{cases}$$

---

## 4. Verify that $f$ is a valid PDF

To be valid, $f(x) \ge 0$ (which is true for $x \in [0, 1]$) and the total area must equal 1:
$$\int_{0}^{1} f(x) \, dx = \int_{0}^{1} \frac{1}{B(2, 18)}x^{2-1}(1-x)^{18-1} \, dx$$
By definition of the Beta function $B(\alpha, \beta) = \int_0^1 t^{\alpha-1}(1-t)^{\beta-1} dt$:
$$\frac{1}{B(2, 18)} \cdot B(2, 18) = 1$$

---

## 5. Compute the expectation $E[X]$

For a Beta distribution, $E[X] = \frac{\alpha}{\alpha + \beta}$:
$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = 0.10$$

---

## 6. Compute the variance $Var(X)$

The formula for variance is $Var(X) = \frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$:
$$Var(X) = \frac{2 \cdot 18}{(2+18)^2(2+18+1)} = \frac{36}{(20^2)(21)} = \frac{36}{400 \cdot 21}$$
$$Var(X) = \frac{36}{8400} = \frac{3}{700} \approx 0.00428$$

---

## 7. Compute $P(X \le 0.10)$

$$P(X \le 0.10) = \int_{0}^{0.10} \frac{1}{B(2, 18)}x(1-x)^{17} \, dx$$

---

## 8. Compute $P(0.05 \le X \le 0.15)$

$$P(0.05 \le X \le 0.15) = \int_{0.05}^{0.15} \frac{1}{B(2, 18)}x(1-x)^{17} \, dx$$

---

## 9. Why the Beta distribution is reasonable

The Beta distribution is ideal for modeling proportions because it is naturally bounded between **0 and 1**. Furthermore, by adjusting $\alpha$ and $\beta$, it can take various shapes (skewed, symmetric, or U-shaped) to represent different manufacturing quality levels.

---

## 10. Interpret Expectation and Variance

* **Expectation ($0.10$):** This represents the average defective rate. On average, 10% of the items produced are expected to be defective.
* **Variance ($\approx 0.0043$):** This represents the consistency of the manufacturing process. A low variance indicates that the defective rate is usually very close to 10%, suggesting a stable process.
