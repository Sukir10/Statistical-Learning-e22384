## 1. Support of $X$

The support of a random variable is the set of values for which the density function is non-zero. For this Triangular distribution:
$$\text{Support} = [4, 10]$$

---

## 2. Probability Density Function $f_X(x)$

For a Triangular distribution with $a=4, b=10, c=6$, the PDF is defined as:

$$f_X(x) = \begin{cases} \frac{2(x-4)}{(10-4)(6-4)} = \frac{x-4}{6}, & 4 \le x < 6 \\ \frac{2(10-x)}{(10-4)(10-6)} = \frac{10-x}{12}, & 6 \le x \le 10 \\ 0, & \text{otherwise} \end{cases}$$

---

## 3. Verify that $f_X$ integrates to 1

The integral over the support is the area of the triangle with base $(b-a)$ and height $h = \frac{2}{b-a}$:
$$\text{Area} = \int_{4}^{10} f_X(x) dx = \int_{4}^{6} \frac{x-4}{6} dx + \int_{6}^{10} \frac{10-x}{12} dx$$
$$\text{Area} = \left[ \frac{(x-4)^2}{12} \right]_{4}^{6} + \left[ -\frac{(10-x)^2}{24} \right]_{6}^{10}$$
$$\text{Area} = \left( \frac{4}{12} - 0 \right) + \left( 0 - (-\frac{16}{24}) \right) = \frac{1}{3} + \frac{2}{3} = 1$$

---

## 4. Compute the expectation $E[X]$

For a Triangular distribution, $E[X] = \frac{a+b+c}{3}$:
$$E[X] = \frac{4 + 10 + 6}{3} = \frac{20}{3} \approx 6.67 \text{ hours}$$

---

## 5. Compute the variance $Var(X)$

The formula for variance is $Var(X) = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}$:
$$Var(X) = \frac{4^2 + 10^2 + 6^2 - (4 \cdot 10) - (4 \cdot 6) - (10 \cdot 6)}{18}$$
$$Var(X) = \frac{16 + 100 + 36 - 40 - 24 - 60}{18} = \frac{28}{18} = \frac{14}{9} \approx 1.56$$

---

## 6. Interpret $a, b,$ and $c$

* **$a = 4$:** The **optimistic** time; the absolute minimum time required if everything goes perfectly.
* **$b = 10$:** The **pessimistic** time; the maximum time allowed or expected if major delays occur.
* **$c = 6$:** The **most likely** time (mode); the duration observed most frequently in similar projects.

---

## 7. Why the Triangular distribution is reasonable


The Triangular distribution is often used in project management (like PERT) because it only requires three intuitive
estimates ($min, max, mode$) from experts. It is helpful when historical data is scarce but expert knowledge about bounds and likelihood is available.

---

## 8. Compute $P(X \le 5)$

Since $5$ is in the first interval $[4, 6)$:
$$P(X \le 5) = \int_{4}^{5} \frac{x-4}{6} dx = \left[ \frac{(x-4)^2}{12} \right]_{4}^{5} = \frac{(1)^2}{12} - 0 = \frac{1}{12} \approx 0.0833$$

---

## 9. Compute $P(5 \le X \le 8)$

Using the Complement/CDF method: $P(X \le 8) - P(X \le 5)$.
To find $P(X \le 8)$, it is easier to use $1 - P(X > 8)$:
$$P(X > 8) = \int_{8}^{10} \frac{10-x}{12} dx = \left[ -\frac{(10-x)^2}{24} \right]_{8}^{10} = 0 - (-\frac{4}{24}) = \frac{1}{6}$$
So, $P(X \le 8) = 1 - \frac{1}{6} = \frac{5}{6}$.
$$P(5 \le X \le 8) = \frac{5}{6} - \frac{1}{12} = \frac{10}{12} - \frac{1}{12} = \frac{9}{12} = 0.75$$

---

## 10. Project Planning and Sensitivity Analysis

* **Project Planning:** Managers use $E[X]$ to create realistic schedules and $Var(X)$ to assess the risk of overrunning deadlines.
* **Sensitivity Analysis:** By varying $a, b,$ or $c$, a planner can see how changes in expert assumptions (e.g., a new delay increasing $b$) impact the overall project timeline and probability of success.
