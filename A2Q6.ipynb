## 1. Compute the shape parameters $\alpha$ and $\beta$

Given $a=8, b=20, c=11$:
* **$\alpha$:**
  $$\alpha = 1 + 4\left( \frac{11-8}{20-8} \right) = 1 + 4\left( \frac{3}{12} \right) = 1 + 4(0.25) = 1 + 1 = 2$$
* **$\beta$:**
  $$\beta = 1 + 4\left( \frac{20-11}{20-8} \right) = 1 + 4\left( \frac{9}{12} \right) = 1 + 4(0.75) = 1 + 3 = 4$$

---

## 2. Support of $X$

The support is the interval of possible durations for the task:
$$\text{Support} = [8, 20]$$

---

## 3. Probability Density Function $f_X(x)$

Substitute $a=8, b=20, \alpha=2, \beta=4$ into the given formula:
$$f_X(x) = \frac{(x-8)^{2-1}(20-x)^{4-1}}{B(2,4)(20-8)^{2+4-1}} = \frac{(x-8)(20-x)^3}{B(2,4) \cdot 12^5}, \quad 8 \le x \le 20$$

---

## 4. Probability Space $(\Omega, \mathfrak{F}, P)$

* **Sample Space ($\Omega$):** The set of all possible durations, $\Omega = [8, 20]$.
* **$\sigma$-algebra ($\mathfrak{F}$):** The Borel $\sigma$-algebra on $[8, 20]$, denoted $\mathcal{B}([8, 20])$.
* **Probability Measure ($P$):** For any event $A \in \mathfrak{F}$, $P(A) = \int_A f_X(x) dx$.

---

## 5. Define $X$ as a random variable

$X$ is a measurable function $X: \Omega \to [8, 20]$ defined by the identity $X(\omega) = \omega$. It represents the continuous duration of the software integration task.

---

## 6. Show that the expectation $E[X]$ is $(a+4c+b)/6$

For a PERT distribution, $E[X]$ is a weighted average where the mode ($c$) is given 4 times the weight of the extremes ($a$ and $b$):
$$E[X] = \frac{8 + 4(11) + 20}{6} = \frac{8 + 44 + 20}{6} = \frac{72}{6} = 12 \text{ days}$$

---

## 7. Compute the variance $Var(X)$

Using the provided approximation:
$$Var(X) = \frac{(E[X]-a)(b-E[X])}{7}$$
$$Var(X) = \frac{(12-8)(20-12)}{7} = \frac{4 \cdot 8}{7} = \frac{32}{7} \approx 4.5714$$

---

## 8. Compute the standard deviation $\sigma$

$$\sigma = \sqrt{Var(X)} = \sqrt{\frac{32}{7}} \approx 2.138 \text{ days}$$

----

## 9. PERT vs. Triangular Distribution

A PERT distribution is often more appropriate than a Triangular distribution because:
1. **Smoothness:** PERT has a smooth, bell-like curve (Beta-based) compared to the "peak" of a Triangle, which better reflects natural randomness.
2. **Mean Bias:** In a Triangular distribution, $E[X] = (8+20+11)/3 = 13$. In PERT, $E[X] = 12$. PERT places more emphasis on the mode $c$, preventing the mean from being pulled too far by an extreme maximum ($b=20$), leading to more realistic estimates in software tasks.

----

## 10. Practical Interpretation

* **$a$ (8 days):** The "best-case" scenario; requires zero technical debt or integration issues.
* **$b$ (20 days):** The "worst-case" scenario; accounts for major bugs or compatibility failures.
* **$c$ (11 days):** The most frequent duration observed in previous integrations.
* **$E[X]$ (12 days):** The planned duration to use in a schedule to account for average risks.
* **$\sigma$ (2.14 days):** The uncertainty; a higher $\sigma$ suggests the task is "high risk" and the actual time could deviate significantly from the plan.
