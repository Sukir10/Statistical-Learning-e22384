##1. Construct a Probability Space $(\Omega, \mathfrak{F}, P)$

 A probability space consists of a sample space, a $\sigma$-algebra, and a probability measure.



   * Sample Space ($\Omega$): The set of all possible outcomes. Since the number of failures can be $0, 1, 2,$ or $3$:



     $$\Omega = \{0, 1, 2, 3\}$$



   * $\sigma$-algebra ($\mathfrak{F}$): Since the sample space is finite, we typically use the power set, which contains all possible subsets of $\Omega$:$$\mathfrak{F} = \mathcal{P}(\Omega)$$



   * Probability Measure ($P$): Defined for each elementary outcome $\omega \in \Omega$ as given:$$P(\{0\}) = 0.50, \quad P(\{1\}) = 0.30, \quad P(\{2\}) = 0.15, \quad P(\{3\}) = 0.05$$

---**bold text**
## 2. Define $X$ as a random variable on this space.

$X$ is a function $X: \Omega \to \mathbb{R}$ that maps each outcome to its numerical value. In this model, $X$ is the identity mapping:
$$X(\omega) = \omega \quad \text{for each } \omega \in \{0, 1, 2, 3\}$$

---

## 3. Determine the probability mass function of $X$.

The probability mass function (PMF) $p_X(k) = P(X = k)$ is:
* $P(X=0) = 0.50$
* $P(X=1) = 0.30$
* $P(X=2) = 0.15$
* $P(X=3) = 0.05$

---

## 4. Compute the expectation $E[X]$.

The expectation is calculated as:
$$E[X] = \sum_{k=0}^{3} k \cdot P(X=k)$$
$$E[X] = (0 \cdot 0.50) + (1 \cdot 0.30) + (2 \cdot 0.15) + (3 \cdot 0.05)$$
$$E[X] = 0 + 0.30 + 0.30 + 0.15 = 0.75$$

---

## 5. Compute the variance $Var(X)$.

First, calculate the second moment $E[X^2]$:
$$E[X^2] = \sum_{k=0}^{3} k^2 \cdot P(X=k)$$
$$E[X^2] = (0^2 \cdot 0.50) + (1^2 \cdot 0.30) + (2^2 \cdot 0.15) + (3^2 \cdot 0.05)$$
$$E[X^2] = 0 + 0.30 + 0.60 + 0.45 = 1.35$$

Then, use the variance formula $Var(X) = E[X^2] - (E[X])^2$:
$$Var(X) = 1.35 - (0.75)^2 = 1.35 - 0.5625 = 0.7875$$

---

## 6. Find $P(X \ge 1)$ and $P(X \ge 2)$.

* $P(X \ge 1) = P(X=1) + P(X=2) + P(X=3) = 0.30 + 0.15 + 0.05 = 0.50$
* $P(X \ge 2) = P(X=2) + P(X=3) = 0.15 + 0.05 = 0.20$

---

## 7. Practical meaning of expectation and variance.

* **Expectation ($E[X]$):** Represents the long-term average number of emergency transformer failures reported per day in this district.
* **Variance ($Var(X)$):** Measures the degree of fluctuation or spread in the number of daily failures around the average.
